```python
from PIL import Image
import os

# Function to convert an image to PNG
def convert_image_to_png(image_path, output_path):
    with Image.open(image_path) as img:
        img.convert('RGBA')  # Convert to RGBA if needed
        img.save(output_path, 'PNG')

# Directory where your images are located
input_directory = "J:/UNC Bootcamp/Web-App-Project/ProjectImages"

# Directory where you want to save the PNGs
output_directory = "J:/UNC Bootcamp/Web-App-Project/ProjectImages/CleanedImages"

# Create output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Loop through each file in the directory
for filename in os.listdir(input_directory):
    if filename.endswith(('.png', '.jpg', '.webp')):  # Add or remove extensions as needed
        file_path = os.path.join(input_directory, filename)
        
        # Create a PNG filename
        output_filename = f"{os.path.splitext(filename)[0]}.png"
        output_path = os.path.join(output_directory, output_filename)
        
        # Convert the image to PNG
        convert_image_to_png(file_path, output_path)
```

```python
from PIL import Image
import os

def remove_bg_and_convert_to_png(image_path, output_path):
    img = Image.open(image_path)
    img = img.convert("RGBA")
    
    # Get data of the image
    data = img.getdata()
    new_data = []
    
    for item in data:
        # Change all white (also shades of whites)
        # pixels to be transparent
        if item[0] in list(range(200, 256)):
            new_data.append((255, 255, 255, 0))  # Change all white (also shades of whites) pixels to transparent
        else:
            new_data.append(item)
    
    img.putdata(new_data)
    img.save(output_path, "PNG")

# Directory where your images are located
input_directory = "J:/UNC Bootcamp/Web-App-Project/ProjectImages/CleanedImages"

# Directory where you want to save the modified PNGs
output_directory = "J:/UNC Bootcamp/Web-App-Project/ProjectImages/CleanedImages"

# Create output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Loop through each file in the directory
for filename in os.listdir(input_directory):
    if filename.endswith(('.png', '.jpg', '.webp')):  # Add or remove extensions as needed
        file_path = os.path.join(input_directory, filename)
        
        # Create a PNG filename
        output_filename = f"{os.path.splitext(filename)[0]}_no_bg.png"
        output_path = os.path.join(output_directory, output_filename)
        
        # Remove background and convert the image to PNG
        remove_bg_and_convert_to_png(file_path, output_path)
```
