```python
import pandas as pd
from pandas import json_normalize
import numpy as np
import json
```

```python
load_df = pd.read_json('all_stores.json')

places_df = []
for i in range(len(load_df['places'])):
    load_df['places'][i] = json_normalize(load_df['places'][i])
    places_df.append(load_df['places'][i])

places_df = pd.concat(places_df, ignore_index=True)
places_df = places_df.drop_duplicates(subset=['_id'])
places_df = places_df.reset_index(drop=True)
places_df
```

```python
load_df = pd.read_json('all_restaurants.json')

restaurants_df = []
for i in range(len(load_df['places'])):
    load_df['places'][i] = json_normalize(load_df['places'][i])
    restaurants_df.append(load_df['places'][i])

restaurants_df = pd.concat(restaurants_df, ignore_index=True)
restaurants_df = restaurants_df.drop_duplicates(subset=['_id'])
restaurants_df = restaurants_df.reset_index(drop=True)
restaurants_df
```

```python
categories_array = restaurants_df['categories'].apply(pd.Series)
categories_array.rename(columns={0: 'r_cat_0',
                                   1: 'r_cat_1',
                                   2: 'r_cat_2',
                                   3: 'r_cat_3'
                                   }, inplace=True)

r_cat_0 = categories_array['r_cat_0'].unique()
print(r_cat_0)
r_cat_1 = categories_array['r_cat_1'].unique()
print(r_cat_1)
r_cat_2 = categories_array['r_cat_2'].unique()
print(r_cat_2)
r_cat_3 = categories_array['r_cat_3'].unique()
print(r_cat_3)

location_array = restaurants_df['location.coordinates'].apply(pd.Series)
location_array.rename(columns={0: 'lat',
                                   1: 'long'
                                   }, inplace=True)


restaurants_final_df = pd.concat([restaurants_df.drop(['categories', 'location.coordinates'], axis=1), categories_array, location_array], axis=1)
restaurants_final_df['score'] = 1
restaurants_final_df['reviews'] = np.nan
restaurants_final_df.head()
```

```python
categories_array = places_df['categories'].apply(pd.Series)
categories_array.rename(columns={0: 'p_cat_0',
                                   1: 'p_cat_1',
                                   2: 'p_cat_2',
                                   3: 'p_cat_3'
                                   }, inplace=True)

p_cat_0 = categories_array['p_cat_0'].unique()
print(p_cat_0)
p_cat_1 = categories_array['p_cat_1'].unique()
print(p_cat_1)
p_cat_2 = categories_array['p_cat_2'].unique()
print(p_cat_2)
p_cat_3 = categories_array['p_cat_3'].unique()
print(p_cat_3)

location_array = places_df['location.coordinates'].apply(pd.Series)
location_array.rename(columns={0: 'lat',
                                   1: 'long'
                                   }, inplace=True)


places_final_df = pd.concat([places_df.drop(['categories', 'location.coordinates'], axis=1), categories_array, location_array], axis=1)
places_final_df['score'] = 1
places_final_df['reviews'] = np.nan
places_final_df.head()
```

```python
places_final_df.rename(columns={'_id': 'id',
                                'name': 'name',
                                'location.type': 'type',
                                'chain.slug': 'slug',
                                'chain.name': 'chain_name',
                                'chain.domain': 'domain',
                                'lat': 'lat',
                                'long': 'long',
                                'score': 'score',
                                'reviews': 'reviews'
                                }, inplace=True)

places_final_df.head()
```

```python
restaurants_final_df.rename(columns={'_id': 'id',
                                'name': 'name',
                                'location.type': 'type',
                                'chain.slug': 'slug',
                                'chain.name': 'chain_name',
                                'chain.domain': 'domain',
                                'lat': 'lat',
                                'long': 'long',
                                'score': 'score',
                                'reviews': 'reviews'
                                }, inplace=True)

restaurants_final_df.head()
```

```python
places_cat_df = places_final_df[['p_cat_0', 'p_cat_1', 'p_cat_2', 'p_cat_3']].copy()
places_cat_df.head()
#for p_cat_0, p_cat_1, p_cat_2, p_cat_3 in zip(p_cat_0, p_cat_1, p_cat_2, p_cat_3
```

```python
import sqlite3

# define connection

connection = sqlite3.connect('../database/radar.db')

cursor = connection.cursor()
places_final_df.to_sql('places', connection, if_exists='replace', index=False)
restaurants_final_df.to_sql('restaurants', connection, if_exists='replace', index=False)
places_cat_df.to_sql('places_cat', connection, if_exists='replace', index=False)
restaurants_cat_df.to_sql('restaurants_cat', connection, if_exists='replace', index=False)
unique_places_cat_df.to_sql('unique_places_cat', connection, if_exists='replace', index=False)
unique_restaurants_cat_df.to_sql('unique_restaurants_cat', connection, if_exists='replace', index=False)

connection.close()
```
